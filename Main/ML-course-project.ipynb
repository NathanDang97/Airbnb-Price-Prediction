### Implement the simple Linear Model

```python
X_train_full = X_train.to_numpy()
y_train_full = y_train.to_numpy()
features_to_drop = set()
```

```python
def correlated_features(X, threshold=0.8):
    correlated_features = set() # set of indices of features
    corr_matrix = np.corrcoef(X.T) # calculate the correlation matrix
    for i in range(len(corr_matrix[0])):
        for j in range(i):
            if np.abs(corr_matrix[i][j]) > threshold:
                correlated_features.add(i)
    return correlated_features
```

```python
def constant_features(X):
    const_features = set()
    for i in range(X.shape[1]):
        if np.std(X[:, i], axis=0) == 0:
            const_features.add(i)
    return const_features
```

```python
def duplicate_features(X):
    dup_features = set()
    for i in range(X.shape[1]):
        for j in range(i+1, X.shape[1]):
            if np.array_equal(X[:,i], X[:,j]):
                dup_features.add(j)
    return dup_features
```

```python
features_to_drop = set()
corr_features = correlated_features(X_train_full)
const_features = constant_features(X_train_full)
dup_features = duplicate_features(X_train_full)
features_to_drop.update(const_features)
features_to_drop.update(dup_features)
```

```python
def preprocess(X, y):
    X = np.delete(X, list(features_to_drop), axis=1)
    return X, y
```

```python
X_train_processed, y_train_full = preprocess(X_train_full, y_train_full)
X_train_processed.shape
```

#### Linear Regression using Gradient Descent and L2-Regularisation

```python
import time

def add_column(X):
    X = np.vstack((np.ones((X.shape[0],)), X.T)).T
    return X

def reg_predict(X_train, theta):
    return np.dot(X_train, theta)

def loss_gradient(X_train, y_train, theta):
    loss_grad = ((reg_predict(X_train, theta)-y_train) * X_train).mean(axis=0)[:, np.newaxis]
    return loss_grad

def loss_gradient_with_reg(X_train, y_train, theta, lambda_=0.0):
    unreg_grad = loss_gradient(X_train, y_train, theta)
    
    reg_term = (lambda_/X_train.shape[0])*theta
    
    reg_term = reg_term.reshape(-1, 1)
    unreg_grad = unreg_grad.reshape(-1, 1)
    reg_term[0] = 0
    
    return unreg_grad + reg_term

def fit(loss_gradient_with_reg, X_train, y_train, theta_init, n_iter = 2000, alpha=0.01):
    theta_current = theta_init.copy()
    for i in range(n_iter):
        theta_current = theta_current - alpha*loss_gradient_with_reg(X_train, y_train, theta_current, 1.0)
    return theta_current

def train(X_train, y_train):
    time_start = time.time()
    
    X_train= add_column(X_train)
    theta_init = np.zeros((X_train.shape[1], 1))
    theta_current = fit(loss_gradient_with_reg, X_train, y_train, theta_init)

    time_end = time.time()
    print(f"{np.round((time_end - time_start) /60, 2)} minutes")
    
    return theta_current

def predict(X_val, theta):
    X_val = add_column(X_val)
    return np.dot(X_val, theta)
```

#### Normal Equation with L2-Regularisation

```python
from numpy.linalg import inv

def search_lambda(X_train_full, y_train_full):
    time_start = time.time()
    
    curr_lambda = 1.0
    final_lambda = 1.0
    min_mse = float('inf')
    avg_curr_mse = 0.0
    
    for i in range(100):
        mse_runs = []
        for i in range(5): # 5-fold cross-validation
            rows = X_train_full.shape[0]
            random_indices = np.random.choice(rows, size=round(0.8*X_train_full.shape[0]), replace=False)
            X_train_curr = np.delete(X_train_full, random_indices, axis=0)
            X_val_curr = X_train_full[random_indices, :]
            y_train_curr = np.delete(y_train_full, random_indices, axis=0)
            y_val_curr = y_train_full[random_indices, :]

            current_theta = train_normal(X_train_curr, y_train_curr, curr_lambda)
            y_pred_curr = predict(X_val_curr, current_theta)
            current_mse = np.square(np.subtract(y_val_curr, y_pred_curr)).mean()
            mse_runs.append(current_mse)
        
        avg_curr_mse = np.mean(mse_runs)
        
        if avg_curr_mse < min_mse:
            min_mse = avg_curr_mse
            final_lambda = curr_lambda
        curr_lambda += 0.1
        
    time_end = time.time()
    print(f"{np.round((time_end - time_start) /60, 2)} minutes")
    
    return min_mse, final_lambda

def train_normal(X_train, y_train, lambda_=0.0):

    lambda_matrix = lambda_ * np.identity(X_train.shape[1])
    lambda_matrix = np.insert(lambda_matrix, 0, 0, axis=0)
    lambda_matrix = np.insert(lambda_matrix, 0, 0, axis=1)
    X_train= add_column(X_train)
    theta = inv(X_train.T @ X_train + lambda_matrix) @ (X_train.T @ y_train)
    
    return theta

def predict(X_val, theta):
    X_val = add_column(X_val)
    return np.dot(X_val, theta)
```
